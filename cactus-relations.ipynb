```python
# define operator q_i = \tau_1 (\tau_2\tau_1) \cdots (\tau_i \cdots \tau_1)
def q1(linext, i):
    for j in range(i,0,-1):
        for k in range(1,j+1):
            linext = linext.tau(k)
    return linext

# define operator q = q_ij = q_{j-1}q_{j-i}q_{j-1}
def q(linext,i,j):
    linext = q1(linext, j-1)
    linext = q1(linext, j-i)
    linext = q1(linext, j-1)
    return linext

# cactus operator
def cactus(linext,i,j,k):
    linext = q(linext,j,k)
    linext = linext.tau(i)
    linext = q(linext,j,k)
    linext = linext.tau(i)
    return linext

# run cactus on all (i,j,l) such that i+1<j<k
def check_cactus(linext):
    n = len(linext)
    for i in range(1,n+1):
        for j in range(i+2,n+1):
            for k in range(j+1,n+1):
                print("(i,j,k)=" + str((i,j,k)))
                print("cactus: " + str(cactus(linext,i,j,k)))
    print()
```

```python
elms = [1,2,3,4]
rels = [[1,3],[1,4],[2,3],[2,4]]
P = Poset((elms, rels), linear_extension=True)
L = P.linear_extensions()

for linext in L:
    print("linext: " + str(linext))
    check_cactus(linext)
```
